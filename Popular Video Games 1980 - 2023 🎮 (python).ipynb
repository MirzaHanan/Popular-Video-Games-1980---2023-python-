# Data Cleaning

```python
#Drop First Column
df.drop(['Unnamed: 0'], axis=1)
```

```python
#get mean of rating column
rating_mean = df["Rating"].mean()

#impute the rating column's mean to all the missing value in its column
df.Rating = df.Rating.fillna(rating_mean)

#check to see if rating column still has missing value
print(df.isnull().sum())
```

# Data Visulisation

```python
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
import seaborn as sns


#First 20 games rating
games = df['Title'].iloc[0:20]

rating = df['Rating'].iloc[0:20]

game_rating = df.groupby(games).size().reset_index(name='Rating')

# Set width and height of figure
plt.figure(figsize=(10,5))

# Add title
plt.title("Top 20 Games")

# Bar chart showing First 20 games rating
sns.barplot(x=rating, y=games, data=game_rating)

# label vertical axis
plt.ylabel("Game Titles")

# label horizontal axis
plt.xlabel("Rating")

# Show the plot
plt.show()
```

```python
# Split the values in the "Genres" column into separate columns

genre_columns = (df['Genres'].str.split(', ', expand=True))

# Clean the genre values
genre_columns = genre_columns.apply(lambda x: x.str.strip("[],"))

# Rename the columns to indicate genre
genre_columns = genre_columns.add_prefix('Genre_')

# Concatenate the cleaned genre columns with the original DataFrame
df = pd.concat([df, genre_columns], axis=1)

# Example: Bar chart of individual genres
genre_counts = genre_columns.stack().value_counts()
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.xticks(rotation=90)
plt.title('Individual Game Genres')
plt.show()
```
